```
import numpy as np 
import random as r
import math
import matplotlib.pyplot as plt

def allProcess():
    fea = 10
    input = 100
    #probability array
    prob = []

    #rounding iteration 
    c = np.random.uniform(-10, 10,(input, fea+1))
    for i in range(input):
      for j in range(fea):
        c[i][j] = round(c[i][j],1)

    #x = each input 
    for x in range(input):
      #ran = randomn class 
      ran = r.randint(0,fea-1)
      c[x][fea] = 0
      #print("random", c[x][ran])
    
      #add prounormalized probaility (i skipped unnormilized log probability)
      prob.append(math.exp(c[x][ran]))
      #print("exp", prob[x])

      #c[x][ran] = random class for x input

      #y = each class 
      for y in range (fea):
        if (y==ran):
          continue

        #formula: sum(max(0, other class value - target value + 1))
        c[x][fea] += max(0, c[x][y] - c[x][ran] + 1)

    #print(c)

    #for i in range(input):
    #  for j in range(fea+1):
    #    print(c[i][j],end=' ')
    #  print()

    #calculation total 
    total = 0
    for l in range(input):
      total += c[l][fea]

    total = total/input
    #print(total)

    #calculation sum of probability of all classes
    sumProb = np.sum(prob)
    #print(sumProb)

    Lc = []
    Li = 0

    #probability
    for p in range(input):
      prob[p] = prob[p]/sumProb
      #print(prob[p])

      #calculation loss function for each class 
      Lc.append(-prob[p]*math.log10(prob[p]))
      #print(Lc[p])

      #calculation loss for input
      Li += Lc[p]


    #calculation total loss
    Lt = Li/input
    #print(Lt)
    return Lt

nResult = []
nRange = []
for i in range(1000):
    nResult.append(allProcess())
    nRange.append(i)

fig, ax = plt.subplots()  # Create a figure containing a single axes.
ax.plot(nRange, nResult)
```
